```python
import cv2
import os
import time
import mediapipe as mp

# Setup constants
DATA_DIR = './dataset2'
folder_name = 'Chotti_Yay'
folder_path = os.path.join(DATA_DIR, folder_name)

# Ensure the target directory exists
os.makedirs(folder_path, exist_ok=True)
print(f"Data will be saved to: {folder_path}")

# Setup MediaPipe Hands
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(
    static_image_mode=False, 
    max_num_hands=2, 
    min_detection_confidence=0.5, 
    min_tracking_confidence=0.5
)
mp_drawing = mp.solutions.drawing_utils

# Initialize video capture
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Camera could not be opened.")
    exit()

print("Press 'S' to start the capture, 'Q' to quit.")
started = False
counter = 0

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error: Can't receive frame (stream end?). Exiting ...")
            break

        if started:
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = hands.process(frame_rgb)
            display_frame = frame.copy()
            
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    mp_drawing.draw_landmarks(
                        display_frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
            
            
            
            img_path = os.path.join(folder_path, f'{counter}.jpg')
            cv2.imwrite(img_path, frame)
            print(f'Captured image saved at: {img_path}')
            counter += 1
            
            elapsed_time = time.time() - start_time
            if elapsed_time > 30:
                print("30 seconds elapsed. Capture finished.")
                break
            
            remaining_time = 30 - int(elapsed_time)
            cv2.putText(display_frame, f'Time left: {remaining_time}s', (10, 50), 
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            cv2.imshow('Hand Tracking', display_frame)
        else:
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = hands.process(frame_rgb)
            display_frame = frame.copy()
            
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    mp_drawing.draw_landmarks(
                        display_frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
                    
            cv2.putText(display_frame, 'Press "S" to start, "Q" to quit', (10, 30), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
            cv2.imshow('Hand Tracking', display_frame)
        
        key = cv2.waitKey(1) & 0xFF
        if key == ord('s') and not started:
            started = True
            start_time = time.time()
            print("Capture started.")
        elif key == ord('q'):
            print("Exiting...")
            break

finally:
    # Clean up resources
    cap.release()
    cv2.destroyAllWindows()
    hands.close()
```
